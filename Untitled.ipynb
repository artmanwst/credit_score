```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline


```

```python
data_set = pd.read_csv("credit_train.csv")
data_set.info()
data_set.shape

```

```python
data_set['Loan Status'].unique()
```

```python
data_set

```

```python

del data_set['Years in current job']

data_set['Credit Score'].fillna(data_set['Credit Score'].median(),inplace=True)
data_set['Months since last delinquent'].fillna(data_set['Months since last delinquent'].median(),inplace=True)
data_set['Maximum Open Credit'].fillna(data_set['Maximum Open Credit'].median(),inplace=True)
data_set['Bankruptcies'].fillna(data_set['Bankruptcies'].median(),inplace=True)
data_set['Annual Income'].fillna(data_set['Annual Income'].median(),inplace=True)
 
data_set['Tax Liens'].fillna(data_set['Tax Liens'].median(),inplace=True)
data_set.dropna(inplace=True)
```

```python
df=pd.get_dummies(data_set,columns=['Loan Status','Term','Home Ownership','Purpose'],drop_first=True)

```

```python

df[0:5]
```

```python

df=df.iloc[:,2:]
```

```python
df.info()
```

```python
x=df.drop('Loan Status_Fully Paid',axis=1)
y=df['Loan Status_Fully Paid']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)



```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf_model = RandomForestClassifier(n_estimators=100, random_state=0, class_weight='balanced')



```

```python
x_train.info()
```

```python
rf_model.fit(x_train, y_train)

y_pred_rf = rf_model.predict(x_test)

accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Точность модели случайного леса:", accuracy_rf)


```
